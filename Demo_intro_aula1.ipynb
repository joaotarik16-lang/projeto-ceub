## Passo 2: Carregar o `pipeline` do `transformers` com o modelo `pysentimiento/bertweet-pt-sentiment` do site huggingface.  

**Exercício:**

* Navegue até o site do Huggingface, pesquise por "analise de sentimento" na aba "modelos", copie o nome de outro modelo, coloque no lugar do modelo abaixo e veja se há diferença nas respostas.

```python
from transformers import pipeline

MODEL = "pysentimiento/bertweet-pt-sentiment"

# pipeline mínimo — sem parâmetros extras
sentiment = pipeline("sentiment-analysis", model=MODEL)
print("Pipeline carregada.")
```

## Passo 3: Criar uma função python simples que imprime na tela o rótulo e a probabilidade associada daquele sentimento.

**Exercício:**
* retire o "str" após o "text: " na função e coloque "int" no lugar. Veja o que acontece e tente entender o motivo do problema causado.
* escreva PRINT() em caixa alta no lugar de um dos "print()" abaixo e veja o que acontece. Tente entender o motivo do problema.

```python
def print_sentiment(text: str):
    result = sentiment(text)[0]  # pega o primeiro (e único) resultado, por isso o [0], pois o python começa a contar a partir do 0
    print("\n\n")
    print("Frase:", text)
    print("Label:", result["label"])
    print("Probabilidade:", result["score"])
```

## Passo 4: utilize a função responsável por chamar o modelo de IA criada anteriormente

**Exercício:**

* execute a célula abaixo quantas vezes quiser, escreva a frase que você quer que seja analisada pela IA e aperte ENTER

```python
frase = input("Digite uma frase: ")
print_sentiment(frase)
```
