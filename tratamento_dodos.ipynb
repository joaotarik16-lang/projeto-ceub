Neste notebook, realizamos a coleta, inspeção e limpeza dos dados de infrações de trânsito do Distrito Federal referentes ao ano de 2025, disponibilizados pelo [Portal de Dados Abertos do DF](https://dados.df.gov.br).

O objetivo é produzir um dataset padronizado e tipado, pronto para análise exploratória.


## 1. Coleta e Consolidação dos Dados

Os dados são disponibilizados mensalmente em arquivos CSV separados. Lemos cada mês diretamente da URL e os consolidamos em um único DataFrame.

## Processamento e Limpeza de Dados (Documentação)


### 1. Padronização de Texto e Valores Nulos
Garante que não haverá duplicatas por causa de letras maiúsculas ou acentos, e trata informações vazias.


```python
# Padroniza os textos para minusculo, sem acento e substitui nulos por 'Nao informado'
import unicodedata
import re

def standardize_text(text):
    if pd.isna(text) or str(text).strip() == '' or str(text).lower() == 'nan': return 'Nao informado'
    text = str(text).lower().strip()
    text = ''.join(c for c in unicodedata.normalize('NFD', text) if unicodedata.category(c) != 'Mn')
    return re.sub(r'\s+', ' ', text)

text_cols = ['tipo_infracao', 'descricao', 'tipo_infrator', 'tipo_veiculo', 'auinf_local_rodovia', 'auinf_local_referencia', 'auinf_local_complemento', 'grav_tipo']
for col in text_cols:
    if col in df.columns: df[col] = df[col].apply(standardize_text)

for col in df.columns:
    if df[col].dtype == 'object': df[col] = df[col].fillna('Nao informado')
```

- Termos como 'AUTOMÓVEL' e 'automovel' passaram a ser idênticos ('automovel').
- Linhas vazias agora indicam de forma explícita que a informação 'Não informado'.


### 2. Conversão de Datas e Criação de Features
Transforma as colunas de data/hora nos seus respectivos formatos nativos e cria a coluna indicando o dia da semana.


```python
# Converte cometimento para Datetime, hora_cometimento para Timedelta e extrai o dia da semana
import numpy as np

if 'cometimento' in df.columns:
    df['cometimento'] = df['cometimento'].replace('Nao informado', np.nan)
    df['cometimento'] = pd.to_datetime(df['cometimento'], format='%d/%m/%Y', errors='coerce')

if 'hora_cometimento' in df.columns:
    def parse_hora(h):
        h = str(h).strip()
        if h.lower() in ['nan', 'nao informado', '', 'nat']: return np.nan
        if len(h) == 5: return h + ':00' # Transforma 14:30 em 14:30:00
        return h
    
    df['hora_cometimento'] = df['hora_cometimento'].apply(parse_hora)
    df['hora_cometimento'] = pd.to_timedelta(df['hora_cometimento'], errors='coerce')

dias_semana = {0: 'segunda-feira', 1: 'terca-feira', 2: 'quarta-feira', 3: 'quinta-feira', 4: 'sexta-feira', 5: 'sabado', 6: 'domingo'}
if 'cometimento' in df.columns:
    df['dia_da_semana'] = df['cometimento'].dt.dayofweek.map(dias_semana).fillna('Nao informado')
```

- O `cometimento` passou a ser o tipo `datetime`.
- A `hora_cometimento` passou a ser o tipo `timedelta`, perfeito para cálculos temporais.
- Criamos a coluna `dia_da_semana` baseada no calendário real.


### 3. Cruzamento de Dados — Valores das Multas

Carregamos a tabela de referência `codigos_valores_multas.csv`, construída a partir dos valores-base do CTB e dos multiplicadores definidos na **Portaria SENATRAN n.º 354/2022** (Anexo N4). O cruzamento é feito pelo código da infração (`tipo_infracao`), gerando a coluna `valor_multa` com o valor em R$ já como `float`.


```python
# Le a tabela de referencia de multas (codigo -> valor em R$)
df_multas = pd.read_csv('codigos_valores_multas.csv', sep=',')
# Renomeia para facilitar o código padrão independentemente do CSV original
df_multas.columns = ['codigo_infracao', 'valor_multa']

# Padroniza o codigo para bater com tipo_infracao (ja em lowercase sem acento)
df_multas['codigo_infracao'] = df_multas['codigo_infracao'].str.strip().str.lower()

# Remove coluna antiga caso ja exista (evita conflito no merge)
if 'valor_multa' in df.columns:
    df.drop(columns=['valor_multa'], inplace=True)

# Merge left: mantem todas as autuacoes, anexa o valor quando ha correspondencia
df = df.merge(df_multas, left_on='tipo_infracao', right_on='codigo_infracao', how='left')
df.drop(columns=['codigo_infracao'], inplace=True)
```

- O cruzamento adicionou a coluna `valor_multa` (tipo `float64`) a cada autuação.
- Infrações sem correspondência na tabela ficam com `NaN` em `valor_multa`.
- Os valores já estão em R$ prontos para uso analítico, sem necessidade de parsing adicional.


### 4. Exportação para Parquet

Salvamos o DataFrame limpo e enriquecido no formato Parquet, que preserva os tipos nativos (`datetime`, `timedelta`, `float64`) e oferece leitura muito mais rápida do que CSV.


```python
# Garante que todas as colunas de texto (object) sejam strings puras para evitar erros no PyArrow
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str)

# Salva o DataFrame final com todas as colunas tipadas corretamente
df.to_parquet('dados_2025.parquet', index=False, engine='pyarrow')

# Verifica o resultado carregando de volta e exibindo tipos e amostra
df_check = pd.read_parquet('dados_2025.parquet')
display(df_check.info())
display(df_check)
```
